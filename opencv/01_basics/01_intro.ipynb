Upload image in gray scale

Image addition

```python
img1 = cv2.imread(path + '/view.jpg')
img2 = cv2.imread(path + '/bear.jpg')
```

```python
add_imgs = cv2.add(src1=img1, src2=img2)
add_imgs.shape
```

```python
cv2_imshow(add_imgs)
```

Image blending

```python
dst = cv2.addWeighted(src1=img1, alpha=0.2, src2=img2, beta=0.7, gamma=0)
cv2_imshow(dst)
```
